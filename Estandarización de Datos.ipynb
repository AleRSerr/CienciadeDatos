# Estandarizacion de Datos con MIN-MAX

La estandarización de datos permite tener los datos de distintas variables en un mismo rango de 0-1 **sin modificar su distribución** ni afectar a una posible correlación.

se pueden escalar los datos de 0-1 utilizando el método MIN-MAX:

\begin{equation}
X'=\frac{X-X~min~}{X~max~-X~min~}
\end{equation}

Se pueden ver los la normalizacion de los datos en la siguiente Tabla:

| Departamento | Expectativa de vida (años) | Pobreza Multidimensional (%) |
| :---- | :-----------: | ------: |
| Antioquia | 71 | 37,59 |

| Nariño | 72 | 45,85 |
| Tolima | 73 | 38,54 |
| Vaupés | 67 | 88,76 |
| Vichada | 67 | 66,77 |

Se pueden observar los siguientes valores MIN-MAX para las variables:

#### Expectativa de vida

| Valor | Departamento | Expectativa de vida (años) | 
| :--- | :---- | :-----------: | 
| **MAX** | Bogotá D.C. | 75 | 
| **MIN** | Vichada | 67 |

| Valor | Departamento | Pobreza Multidimensional (%) |
| :---- | :---- | :-----------: | 
| **MAX** | Vaupés | 88,76 |

Aplicando la estandarizacion de datos, nuestra nueva tabla se veria de la siguiente manera:

tomando en cuenta lo anterior se puede generar el siguiente grafico de dispersión:

1. Se importa primero matplotlib
2. se ingresan los datos para generar el plano de dispersión

```python
import matplotlib.pyplot as plt
```

```python
x = [0.50,1.00,0.63,0.75,0.00,0.00]
y = [0.39,0.00,0.46,0.37,1.00,0.72]
plt.scatter(x,y,label="Departamentos",marker="*",s=70,color="blue")
plt.xlabel("Expectativa de Vida\nEstandarizado")
plt.ylabel("Pobreza Multidimensional\nEstandarizado")
plt.title("Plano de Dispersión\nExpectativa de vida vs. Pobreza Multidimensional")
plt.legend()
plt.show()
```
